```python
# 处理数据以便多项式回归
def poly_features(X, degree):
    """生成多项式特征矩阵"""
    X = np.array(X).reshape(-1, 1)  # 确保是列向量
    features = [np.ones_like(X)]  # 包含常数项
    
    for i in range(1, degree+1):
        features.append(X**i)
    
    return np.hstack(features)

poly_X = poly_features(X,5)
poly_X.shape
```

```python
# 多项式回归
def Reg_CostFunction(X:np.ndarray,y:np.ndarray,Theta:np.ndarray,lambda_:float)->float:
    y.reshape(-1.1)
    m=y.shape[0]
    Theta.reshape(-1,1)
    error = np.sum(np.power(X @ Theta - y,2))
    reg = lambda_ / (2.0*m)*np.sum(np.power(Theta[1:],2))
    return error + reg
```
